```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = '/content/drive/MyDrive/WASSUP/Sequence_Data/drug_data.csv'
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
# See types of each column
print(df.info())
```

```python
# Drop unused column
df.drop('Unnamed: 30', axis=1, inplace=True)
```

```python
# Check shape of dataframe
print(f"Shape of data: {df.shape}\n")
```

```python
# Check first 5 values
display(df.head())
```

```python
# Check last 5 values
display(df.tail())
```

```python
# Check missing values
print("Missing values in data")
print(df.isna().sum())
```

```python
# Check duplicated values
print(f"Duplicated values in data: {len(df[df.duplicated(subset='품목일련번호')])}")
display(df[df.duplicated(subset='품목일련번호')])
```

```python
# Transform type from object to numeric
df['크기장축'] = pd.to_numeric(df['크기장축'], errors='coerce')
df['크기단축'] = pd.to_numeric(df['크기단축'], errors='coerce')
df['크기두께'] = pd.to_numeric(df['크기두께'], errors='coerce')
print(df['크기장축'].dtypes)
print(df['크기단축'].dtypes)
print(df['크기두께'].dtypes)
```

```python
# Check statistic information
df[['크기장축', '크기단축', '크기두께']].describe()
```

```python
# Copy some columns
df['long_axis'] = df['크기장축']
df['short_axis'] = df['크기단축']
df['thickness'] = df['크기두께']
```

```python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(df[['long_axis', 'short_axis', 'thickness']].corr(numeric_only=True), annot=True, fmt='.4f', cmap='RdBu_r', center=0, annot_kws={"fontsize":12})
plt.title('Correlation Heatmap', fontsize=20)
plt.show()
```

```python
# Draw scatter plot
plt.scatter(df['long_axis'], df['short_axis'], s=5)
plt.title('Scatter plot of long axis and short axis')
plt.xlabel('long axis')
plt.ylabel('short axis')
plt.xlim([0, 45])
plt.ylim([0, 35])
plt.grid()
plt.show()
```

```python
# Draw box plot to see outliers
df.boxplot(column=['long_axis','short_axis', 'thickness'])
plt.ylim([0, 50])
plt.show()
```
